```python
from sklearn.tree import DecisionTreeClassifier

# Features: [Weight (g), Texture (0=Smooth,1=Rough), Color (0=Red,1=Orange,2=Yellow)]
features = [
    [150, 0, 0],  # Apple
    [170, 0, 0],  # Apple
    [130, 1, 1],  # Orange
    [120, 1, 1],  # Orange
    [160, 0, 0],  # Apple
    [145, 0, 0],  # Apple
    [155, 0, 0],  # Apple
    [180, 1, 1],  # Orange
    [135, 1, 1],  # Orange
    [200, 1, 1],  # Orange
    [100, 0, 2],  # Banana
    [110, 0, 2],  # Banana
    [95,  0, 2],  # Banana
    [105, 0, 2],  # Banana
]

labels = [
    'apple', 'apple', 'orange', 'orange',
    'apple', 'apple', 'apple', 'orange',
    'orange', 'orange', 'banana', 'banana',
    'banana', 'banana'
]

# Train the Decision Tree
clf = DecisionTreeClassifier()
clf.fit(features, labels)

# Predict new fruits
predictions = clf.predict([
    [150, 0, 0],   # A 150g smooth red fruit
    [130, 1, 1],   # A 130g rough orange fruit
    [100, 0, 2],   # A 100g smooth yellow fruit
    [180, 1, 1],   # A 180g rough orange fruit
    [115, 0, 2],   # A 115g smooth yellow fruit
    [30, 0, 1]
])

print(predictions)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(11,6))
plot_tree(clf, feature_names=["weight", "texture","color"], class_names=["apple", "orange", "banana"], filled=True)
```
