In Python, the pivot function in pandas is used to reshape a DataFrame by converting it from a long-form to a wide-form structure. It allows you to specify the index, columns, and values for the new DataFrame. The function does not support data aggregation, and if there are any duplicate entries for the specified index and column combinations, it raises a ValueError.
 For example, you can use df.pivot(index="lev1", columns=["lev2", "lev3"], values="values") to reshape the DataFrame.

```python
pv = flights.pivot(index="month", columns='year', values='passengers')
```

```python
pv
```

```python
sns.heatmap(pv, annot=True, fmt='d', cmap="YlGnBu")
```
