```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("iris_train.csv")
data.head()
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
print(x,y)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
test_data = pd.read_csv("iris_test.csv")
test = test_data.iloc[:, :-1]
matching_data = test_data.iloc[:, -1]
predictions = model.predict(test)
predictions
```

```python
results = pd.DataFrame({
    "actual":   matching_data,
    "predicted": predictions
})

results["correct"] = results["actual"] == results["predicted"]
results["pct_correct"] = results["correct"].astype(int) * 100  # 100 or 0

print(results)

print("\nOverall accuracy: {:.2f}%".format(results["correct"].mean() * 100))
```
