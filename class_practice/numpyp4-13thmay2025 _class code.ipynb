```python
import numpy as np
import matplotlib.pyplot as plt

# Create a small 3x3 RGB image
img = np.zeros((3, 3, 3), dtype=np.uint8)  # Initialize with all zeros (black)

# --- Pixel Locations ---
# (row, column)
top_left     = (0, 0)
top_middle   = (0, 1)
top_right    = (0, 2)
middle_left  = (1, 0)
middle_middle= (1, 1)
middle_right = (1, 2)
bottom_left  = (2, 0)
bottom_middle= (2, 1)
bottom_right = (2, 2)

# --- Color Definitions ---
red   = [255, 0, 0]
green = [0, 255, 0]
blue  = [0, 0, 255]
white = [255, 255, 255]
black = [0, 0, 0]
magenta = [255, 0, 255]
cyan = [0, 255, 255]
yellow = [255, 255, 0]

# --- Manipulating Pixels ---

# First Row
img[top_left]     = red       # Top-left pixel to red
img[top_middle]   = black     # Top-middle pixel to green
img[top_right]    = black      # Top-right pixel to blue

# Second Row
img[middle_left]  = black    # Middle-left pixel to yellow
img[middle_middle]= green   # Center pixel to magenta
img[middle_right] = black      # Middle-right pixel to cyan

# Third Row
img[bottom_left]  = black     # Bottom-left pixel to white
img[bottom_middle]= black#[128, 128, 128] # Gray
img[bottom_right] = blue     # Bottom-right pixel to black

# --- Display the Image ---
plt.imshow(img)
plt.title("Manipulated 3x3 RGB Image")
plt.axis('off')  # Hide axes
plt.show()

# --- Explanation ---
# print("Explanation:")
# print("- `img[row, column] = [R, G, B]` sets the color of the pixel at the specified row and column.")
# print("- `R`, `G`, and `B` are the red, green, and blue color components, ranging from 0 to 255.")
# print("- The pixel locations (top_left, middle_middle, etc.) are defined as tuples (row, column) for clarity.")
# print("- You can change the colors and pixel locations to experiment with different patterns.")
```

# Let's Create a 5×5 checkerboard pattern

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 5x5 image with 3 channels (RGB)
checkerboard = np.zeros((5, 5, 3), dtype=np.uint8)
checkerboard
```

```python
plt.imshow(checkerboard)
```

```python
black = [0,0,0]
white = [255,255,255]

for i in range(5):
    for j in range(5):
        if (i + j) % 2 == 0:
            checkerboard[i,j] = white
        else:
            checkerboard[i,j] = black
plt.imshow(checkerboard)
```

When working with image processing in Python, both PIL (Pillow) and Matplotlib have their own strengths. Here’s a comparison to help you decide which one to use based on your needs:

Use PIL (Pillow) when:
Loading and Saving Images: It provides straightforward methods to open and save images in various formats.
Basic Image Manipulations: You can easily resize, crop, rotate, and apply filters.
Image Creation: If you need to create images from scratch or modify pixel data directly.

Use Matplotlib when:
Displaying Images: It offers convenient functions to display images inline, especially in Jupyter notebooks.
Plotting Images: Useful for overlaying plots, histograms, or other data visualizations on images.
Image Analysis: When you want to visualize data distributions or perform analysis on image data.

Recommendation:
For general image processing tasks (loading, saving, basic manipulation): Use PIL.
For visualization (showing images, plots, histograms): Use Matplotlib.

Understanding Image Loading
PIL (Image.open):

When you use Image.open, it opens the image file and returns a PIL Image object.
This object can be converted to a NumPy array using np.array(), but plt.imshow can handle the PIL Image directly.
Matplotlib (plt.imread):

plt.imread reads the image file and directly returns it as a NumPy array.
You can pass this array directly to plt.imshow.
When to Convert to NumPy Array
You may need to convert to a NumPy array in the following situations:

NumPy Operations: If you want to perform operations that require NumPy functions (like mathematical operations on pixel values), you will need the image as a NumPy array.

Manipulations: If you are modifying pixel values (e.g., applying filters, changing colors), it's often easier to work with the NumPy array representation.

Compatibility with Other Libraries: Some libraries or functions may require images to be in NumPy array format for processing.

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    img = plt.imread('/kaggle/input/kittydata/kitty.png')
except FileNotFoundError:
    print("Image file not found.")
    exit()

# Check if the image is in uint8 format
if img.dtype != np.uint8:
    img = (img * 255).astype(np.uint8)  # Convert to uint8 if needed

# Invert the image
inverted_img = 128 - img  # For uint8 images

# Display the original and inverted images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(inverted_img)
plt.title("Inverted Image")
plt.axis('off')

plt.show()
```
