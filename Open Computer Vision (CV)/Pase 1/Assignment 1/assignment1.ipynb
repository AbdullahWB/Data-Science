```python
import cv2
import matplotlib.pyplot as plt
```

```python
class Image:
    def __init__(self,ImageName,NewImageName):
        self.ImageName = ImageName
        self.NewImageName = NewImageName

    def LoadImage(self):
        image = cv2.imread(self.ImageName)
        if image is None:
            print(f"Error: Could not load image '{self.ImageName}'. Please check the file path and name.")
        return image

    def GryImage(self):
        image = self.LoadImage()
        if image is not None:
            grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            return grayImage
        return None 

    def showGrayImage(self):
        image = self.GryImage()
        if image is not None:
            plt.imshow(image, cmap='gray')
            plt.title("Grayscale Image (Matplotlib)")
            plt.axis('off')
            plt.show()
        else:
            print("Cannot show grayscale image: Original image not loaded or grayscale conversion failed.")

    def showNormalImage(self):
        image = self.LoadImage()
        if image is not None:
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            plt.imshow(image_rgb)
            plt.title("Original Image (Matplotlib)")
            plt.axis('off')
            plt.show()
        else:
            print("Cannot show normal image: Original image not loaded.")

    def saveImage(self):
        image = self.GryImage()
        if image is not None:
            try:
                write_success = cv2.imwrite(self.NewImageName, image)
                if write_success:
                    print(f"Image saved successfully as '{self.NewImageName}'")
                else:
                    print(f"Image could not be saved as '{self.NewImageName}'. Check permissions or path.")
            except Exception as e:
                print(f"An error occurred while saving the image: {e}")
        else:
            print("Cannot save image: Grayscale image not available.")
```

```python
ImageName = input("Enter your Image name: ")
NewImageName = input("Enter Your new Image Name: ")

newImage = Image(ImageName, NewImageName)
newImage.showNormalImage()
newImage.showGrayImage()
newImage.saveImage()

print("Script execution finished.")
```
