```python
# Read the new image using OpenCV
newImage = cv2.imread("NewTanjiro.png")
```

```python
# --- Use Matplotlib for displaying the image ---
if newImage is not None: # Always good to check if image was loaded
    # OpenCV reads images in BGR format by default.
    # Matplotlib expects RGB. Convert it for correct colors.
    newImage_rgb = cv2.cvtColor(newImage, cv2.COLOR_BGR2RGB)

    plt.imshow(newImage_rgb)
    plt.title("Image Show (Matplotlib)")
    plt.axis('off') # Hide axes for cleaner image display
    plt.show() # Display the plot
else:
    print("Error: Could not load NewTanjiro.png")

print("done")
```
