```python
import cv2
import matplotlib.pyplot as plt
```

```python
bgimage = cv2.imread("cannyImage.jpg")
image = cv2.cvtColor(bgimage, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(bgimage, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
ret, thresh_img = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Image Transformations', fontsize=16)
axes = axes.flatten()

axes[0].imshow(gray , cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(thresh_img, cmap='gray')
axes[1].set_title("thresh_img Image")
axes[1].axis('off')
```
