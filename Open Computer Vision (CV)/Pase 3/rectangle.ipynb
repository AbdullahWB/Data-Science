```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("catgojop.jpg")
plt_image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
pt1 = (100, 200) #left , top
pt2 = (400, 500) # right, bottom
color = (255, 0, 0)
thick = 4

cv2.rectangle(plt_image_rgb, pt1, pt2, color, thick)
plt.imshow(plt_image_rgb)
```
