```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("catgojop.jpg")
plt_image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
pt1 = (100,100)
pt2 = (300, 200)
color = (255, 0, 0)
thickness = 4

cv2.line(plt_image_rgb, pt1, pt2, color, thickness)
plt.imshow(plt_image_rgb)
```
