```python
import cv2
import matplotlib.pyplot as plt
```

```python
Bgrimage = cv2.imread("noiceImage.png")
image = cv2.cvtColor(Bgrimage, cv2.COLOR_BGR2RGB)

blurred = cv2.medianBlur(image, 11)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Image Transformations', fontsize=16)
axes = axes.flatten()

axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(blurred)
axes[1].set_title("blur Image")
axes[1].axis('off')
```
