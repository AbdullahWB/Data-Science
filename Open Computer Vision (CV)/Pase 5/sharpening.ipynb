```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
bgimage = cv2.imread("blurry-low-quality-portrait-photo.png")
image = cv2.cvtColor(bgimage, cv2.COLOR_BGR2RGB)

sharpen_kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
])

sharpened = cv2.filter2D(image, -1 , sharpen_kernel)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Image Transformations', fontsize=16)
axes = axes.flatten()

axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(sharpened)
axes[1].set_title("sharpened Image")
axes[1].axis('off')
```
