```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("girlImage2.jpg")
```

```python
(h, w) = image.shape[:2]
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 90, 1.5)
rotated = cv2.warpAffine(image, M, (w,h))
```

```python
plt_image_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
plt.imshow(plt_image_rgb)
```
