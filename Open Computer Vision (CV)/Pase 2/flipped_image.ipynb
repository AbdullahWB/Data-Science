```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("girlImage2.jpg")
```

```python
plt_image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
flipped_h = cv2.flip(plt_image_rgb, 1)
flipped_v = cv2.flip(plt_image_rgb, 0)
flipped_b = cv2.flip(plt_image_rgb, -1)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8)) 
fig.suptitle('Image Transformations', fontsize=16)
axes = axes.flatten()
axes[0].imshow(plt_image_rgb)
axes[0].set_title("Original Image")
axes[0].axis('off') \

axes[1].imshow(flipped_h)
axes[1].set_title("Flipped Horizontally")
axes[1].axis('off')

axes[2].imshow(flipped_v)
axes[2].set_title("Flipped Vertically")
axes[2].axis('off')

axes[3].imshow(flipped_b)
axes[3].set_title("Flipped Both Ways")
axes[3].axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle
plt.show()
```
